```python
import re

def convert(obj):
    l = []
    # Extract the words within parentheses from the paragraph
    words_in_brackets = re.findall(r'\((.*?)\)', obj)

    # Print the extracted words
    for word in words_in_brackets:
        l.append((word.strip()))
    return l

```

```python
df['itesms'] = df['ingredients'].apply(convert)
```

```python
df['items'] = [str(a).replace("/",'')for a in df['itesms']]
```

```python
df.drop(['itesms'], axis=1,inplace=True)
```

```python
df.drop(['ingredients','instructions','description'], axis=1,inplace=True)
```

```python
df['diet'] = [a.replace(" ", '') for a in df['diet']]
```

```python
df['diet'] = [a.replace("(Sattvic)", '') for a in df['diet']]
```

```python
df['diet'] = [a.lower() for a in df['diet']]
```

```python
df['course'] = [a.lower() for a in df['course']]
df['cuisine'] = [a.lower() for a in df['cuisine']]
```

```python
df['cuisine'] = [a.replace("recipes","") for a in df['cuisine']]
```

```python
df['cuisine'] = [a.replace(" ","") for a in df['cuisine']]
df['course'] = [a.replace(" ","") for a in df['course']]
```

```python
df['items'] = [a.replace(" ","") for a in df['items']]
```

```python
df['cuisine'] = df['cuisine'].apply(lambda x: x.split())
df['course'] = df['course'].apply(lambda x: x.split())
df['diet'] = df['diet'].apply(lambda x: x.split())
```
